### Multiple Linear Regression

```python
 import pandas as pd
```

```python
dataset = pd.read_csv("employee_income.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.rename(columns={"income":"salary"}, inplace = True)
```

```python
dataset.head()
```

```python
dataset.shape
```

No need to scale as the age and experience both are along the same lines. 

Now create a graph for age vs salary and experience vs salary to see the nature

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(data = dataset, x_vars=["experience","age"], y_vars=["salary"])
plt.show()
```

```python
sns.heatmap(data=dataset.corr(),annot=True) #correlation also helps us know about the linearity and annot True also gives us the number
```

```python
x = dataset.iloc[:,:-1] #leaving salary as that is the y
y = dataset["salary"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression #we use the same for multiple linear regression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
mlr = LinearRegression()
mlr.fit(x_train, y_train)
```

```python
mlr.score(x_test, y_test) *100
```

```python
# We see an accuracy of 93 % which is pretty good.
# y= m1x1+m2x2+C
```
