```python
dataset.columns
```

We need to make a list of list of what customers buy. [[a,b], [a,c,e]..] We do this to removoe the NaN values. NaN are the floating content. We will separate the string values and the floating values

```python
dataset.shape
```

```python
# we want to run the loop up until the last row. technically 700 will be empty as it starts at 0-699 but count is 700
market = []
for i in range(0,dataset.shape[0]):
    cus = []
    for j in dataset.columns:
        if type(dataset[j][i]) == str: #if str, add it to customer
            cus.append(dataset[j][i])
        market.append(cus) #add the customer list to market
```

```python
# now we use this market for analysis
market
```

```python
# importing to use the Counter function
import collections
```

```python
# counter function helps us count the occurance of the str/char. Example below
collections.Counter(["a","a","a","s","s", "d"])
```

```python
# To use Counter, we would have to convert the 2-D list to a single list, for which we use numpy
```

```python
l = []
for i in market:
    for j in i:
        l.append(j)
```

```python
# gives count of occurances of each element. Can use p.keys and .values() to see the result.
p = collections.Counter(l)
```

```python
# creating a dictionary with the keys and values
d = {"Item Name":p.keys(), "values":p.values()}
```

```python
#pd.DataFrame(d).sort_values(by=["values"], ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr = TransactionEncoder()
tr.fit(market)
```

```python
df = pd.DataFrame(tr.transform(market), columns=tr.columns_)
#df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(df, min_support=0.05, use_colnames=True).sort_values(by =["support"])
```
