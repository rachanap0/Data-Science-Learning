### Cross Validation

```python
import pandas as pd
```

```python
dataset = pd.read_csv("/Users/rachanapandey/Desktop/study_materials/Data Science/Dataset/Placement.csv")
dataset.head()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["package"]
```

```python
new_data = dataset.head(10)
```

```python
x_new = new_data.iloc[:,:-1]
y_new = new_data["package"]
```

```python
from sklearn.model_selection import LeaveOneOut, LeavePOut, KFold, StratifiedKFold
```

```python
#using leave one out. Takes one at a time for testing and the rest for training.
lo=LeaveOneOut()
for train, test in lo.split(x_new,y_new):
    print(train, test)
```

```python
#using leave p out. Takes p at a time for testing and p for training.
lp = LeavePOut(p=2)

for train, test in lp.split(x_new, y_new):
    print(train, test)
```

```python
#using k-fold which uses k equal split
kf = KFold(n_splits=5)

for train, test in kf.split(x_new, y_new):
    print(train, test)
```

#using stratified k-fold to cross validate
sk = StratifiedKFold(n_splits=5)

for train, test in sk.split(x_new, y_new):
    print(train, test)
