### DBSCAN Clustering
- used for non-linear dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
# Make_moon is a 2D data and gives us two outputs, so we use x and y
x,y = make_moons(n_samples=250, noise=0.05)
```

```python
 df = {"data1": x[:,0], "data2": x[:,1], "output": y }
```

```python
dataset = pd.DataFrame(df)
```

```python
dataset.head()
```

```python
# we can see this is non-linear
sns.scatterplot(data = dataset, x="data1", y="data2")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
# eps is the epsilon i.e. the radius.
db = DBSCAN(eps=0.2, min_samples=5)
dataset["Predict"] = db.fit_predict(dataset)
```

```python
sns.scatterplot(data=dataset, x="data1", y="data2", hue = "Predict")
plt.show()
```
