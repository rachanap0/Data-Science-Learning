### HyperParameter Tuning

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

```python
dataset = pd.read_csv("/Users/rachanapandey/Desktop/study_materials/Data Science/Dataset/salary.csv")
dataset.head()
```

```python
x = dataset.iloc[:,:-1]
y = dataset['Salary']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
dt = DecisionTreeRegressor(criterion='absolute_error',max_depth=10, splitter= 'random')
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)*100
```

```python
dt.score(x_train, y_train)*100
```

### GridSearchCV

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
df = {"criterion":["squared_error", "friedman_mse", "absolute_error", "poisson"],
    "splitter":["best", "random"],
    "max_depth":[i for i in range(2,20)]}
```

```python
gd = GridSearchCV(DecisionTreeRegressor(), param_grid=df)
gd.fit(x_train, y_train)
```

```python
gd.best_params_
```
