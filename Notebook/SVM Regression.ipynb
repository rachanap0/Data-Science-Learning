### Support Vector Machine Regression
- Used for both regression and classification
- Unlike for SVM Classification, tries to find the two support vectors in such a way that they have min distance between and also includes other points in the hyperplane region.
- Disadvantage: Some points might be beyond the decision line and might think of it as an outlier.
- Can be used for both linear and non-linear datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVR
```

```python
dataset = pd.read_csv("/Users/rachanapandey/Desktop/study_materials/Data Science/Dataset/placement.csv")
dataset.head()
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(data = dataset, x= "cgpa", y= "package")
plt.show()
```

```python
x = dataset[['cgpa']]
y = dataset['package']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
svm = SVR(kernel='linear')
svm.fit(x_train, y_train)
```

```python
svm.score(x_test, y_test)*100
```

```python
svm.score(x_train, y_train)*100
```

```python
sns.scatterplot(x="cgpa", y="package", data=dataset)
plt.plot(dataset["cgpa"], svm.predict(x), color = "red")
plt.show()
```
