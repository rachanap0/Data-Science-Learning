### Agglomerative Hierarchial Clustering
Used only for linearly separable data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/Users/rachanapandey/Desktop/study_materials/Data Science/Dataset/iris raw.csv")
dataset.head(2)
```

# check if the data is linearly separable
sns.pairplot(data= dataset)
plt.show()

```python
# making dendogram to find the number of clusters
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20, 10))
sc.dendrogram(sc.linkage(dataset, method='single', metric='euclidean'))
plt.savefig("/Users/rachanapandey/Desktop/study_materials/Data Science/Graphs/Agglomerative.jpg")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ag = AgglomerativeClustering(n_clusters=2, linkage='single')
ag.fit_predict(dataset)
```

```python
dataset['predict'] = ag.fit_predict(dataset)
```

```python
dataset.head(2)
```

```python
sns.pairplot(data=dataset, hue = "predict")
plt.show()
```
