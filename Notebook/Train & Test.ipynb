Now we divide our dependent and independent variables/features. I.E. we want the input and output data. As we can see, the Hous Price is the output of the other variables which is why we want to remove it from input data and store as output data using iloc i.e. integer location-based indexing.

```python
from sklearn.model_selection import train_test_split
```

```python
input_data = dataset.iloc[:,:-1]
output_data = dataset["House_Price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size =0.25 ) # i.e. input data, output data
#usually keep 20-25% test data
```
