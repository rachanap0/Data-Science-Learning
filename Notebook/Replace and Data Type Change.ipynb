```python
import pandas as pd
```

```python
dataset = pd.read_csv("Loan.csv")
dataset.head()
```

Here, dependents is mixed type

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset["Dependents"].value_counts()
```

We fill it with mode and not mean as it is discrete

```python
dataset.fillna({"Dependents":dataset["Dependents"].mode()[0]}, inplace = True)
```

```python
dataset.isnull().sum()
```

### Let's replace 3+ with only 3 as there is no any 3 in the dataset

```python
dataset["Dependents"].replace("3+","3", inplace = True)
```

Need to convert the object type to int or float

```python
dataset["Dependents"] = dataset["Dependents"].astype("int64")
```
