### Decision Tree Regression
- We use it for non-linear dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/Users/rachanapandey/Desktop/study_materials/Data Science/Dataset/employee_income.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["income"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
dt = DecisionTreeRegressor(max_depth=17)
dt.fit(x_train,y_train)
```

```python
dt.score(x_test, y_test)*100
```

```python
dt.score(x_train, y_train)*100
```

```python
plot_tree(dt)
plt.show()
```

```python
for i in range(1,30):
    dt2 = DecisionTreeRegressor(max_depth=i)
    dt2.fit(x_train, y_train)
    print(i, dt2.score(x_train, y_train)*100, dt2.score(x_test, y_test)*100)
```
