### Confusion Matrix
Sometimes even tho the model has a really good accuracy like 90%, 95%.. it might still get rejected. 
Confusion matrix helps us figure out those scenarios and helps us improve the model.

TN: True -ve (0,0)

TP: True +ve (1,1)

FN: False -ve (1,0) Type II error

FP: False +ve (0,1) Type I error

Depending on the scenario, either type I or type II is the reason for rejection.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("placement.csv")
dataset.head()
```

```python
dataset.rename(columns={"resume_score":"score"}, inplace= True)
```

```python
dataset.head()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test) *100
```

### making Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score
```

```python
cm =confusion_matrix(y_test, lr.predict(x_test))
```

```python
sns.heatmap(cm, annot = True)
```

```python
precision_score(y_test, lr.predict(x_test))*100
```

```python
recall_score(y_test, lr.predict(x_test))*100
```

```python
f1_score(y_test, lr.predict(x_test))*100
```
