```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
dataset = pd.read_csv("Loan.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
en_data = [['Rural','Semiurban','Urban']]
```

```python
oe = OrdinalEncoder(categories= en_data)
```

```python
dataset["Property_Area"] = oe.fit_transform(dataset[["Property_Area"]])
```
