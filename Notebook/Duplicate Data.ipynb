```python
import pandas as pd
```

```python
#creating a dataset with duplicate data
data = {"name":["a","b","c","d","a","c"], "eng":[8,7,5,8,8,5],"hindi":[2,3,4,5,2,7]}
```

```python
df = pd.DataFrame(data)
df
```

only row 0 and 4 are duplicates

```python
df.duplicated()
```

df["duplicates"] = df.duplicated()
df

```python
df.drop_duplicates(inplace = True)
```
