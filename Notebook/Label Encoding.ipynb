### Label Encoding
Used for Nominal Data i.e. data that do not have any sequence or order.

```python
import pandas as pd
```

```python
df = pd.DataFrame({"name":["wscube","cow","cat","dog","black"]}) 
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df["en_name"]=le.fit_transform(df["name"]) #storing the transformed numeric data in a new col called en_name
```

```python
df
```

```python
dataset = pd.read_csv("Loan.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["Property_Area"].unique()
```

```python
la =LabelEncoder()
la.fit(dataset["Property_Area"])
```

```python
dataset["Property_Area"] = la.transform(dataset["Property_Area"])
#we could have done fit_transform at once.
```

```python
# we can see that each of the property Area now has a number instead of category 
dataset["Property_Area"].unique()
```
