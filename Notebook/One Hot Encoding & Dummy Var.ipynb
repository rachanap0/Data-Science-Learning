### Encoding
=> Converting categorical data into numerical data

### One Hot Encoding
=> We use One Hot encoding when a certain column has only limited amount of data like 'yes/no', 'male/female', 'semi/urban/rural'...

=> To understand: if we have a column called Gender, it will divide it to Gender_Male and Gender_Female and for every row if the val is male it gives 0 1 else 1 0 for female and that is how our model can understand this column.

```python
 import pandas as pd
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(24)
```

```python
dataset.isnull().sum()
```

```python
dataset.select_dtypes(include="object").columns
```

```python
#finding mode of gender
dataset["Gender"].mode()[0]
```

```python
#filling the empty value of gender with the mode
dataset.fillna({"Gender":dataset["Gender"].mode()[0]}, inplace = True)
```

```python
dataset.head(24)
```

```python
dataset["Married"].mode()[0]
```

```python
dataset.fillna({"Married":dataset["Married"].mode()[0]}, inplace = True)
```

```python
dataset.isnull().sum()
```

### Using get Dummies method in One Hot Encode 

```python
en_data = dataset[["Gender","Married"]]
```

```python
pd.get_dummies(en_data)
```

```python
#gave result in boolean so we would have to convert this to numerical. We can use Scikit for easier approach.
pd.get_dummies(en_data).info()
```

### Scikit-learn (One Hot Encoder)

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder( )
arr = ohe.fit_transform(en_data).toarray() #saving into an array
```

```python
pd.DataFrame(arr, columns=["Gender_Female","Gender_Male","Married_No","Married_Yes" ])
```

```python
#since now we have too many columns, we can use drop="first" to drop the first column of each of the Data. 
#We can do this as the value is still 0 and 1 so we can figure out if a row is 0 for male, it was 1 for female.
```

```python
ohe = OneHotEncoder(drop="first")
arr = ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(arr, columns=["Gender_Male","Married_Yes"])
```
