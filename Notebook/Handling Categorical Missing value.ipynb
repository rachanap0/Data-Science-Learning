### Handling Missing Values (Categorical)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan.csv")
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
#object dtype are categorical data
dataset.info()
```

## Two ways to fill data:

1) Backward Filling: It will fill the empty val in a column with the data behind it. For example if space A11 is empty and A21 has a value. A11 gets the value of A21.
Uses: dataset.fillna(method="bfill")

2) Forward Filling: It will fill the empty val in a column with the data before it. For example if space A11 is filled and A21 has is empty. A21 gets the value of A11. But if A11 was empty, it would still be empty as there is no data above it.
Uses: dataset.fillna(method="ffill")

Note: Since no axis is mentioned here, it will be axis=0 i.e. the forward and backward filling will be done using row.

we can set axis=1 to use the column.

3) Mode: fill missing data with the most repeating data i.e the mode
   

```python
#Backward filling, using the row
dataset.fillna(method="bfill")
```

```python
dataset.fillna(method="ffill", axis =1)
```

```python
dataset.fillna(method="bfill", axis =1)
```

```python
mode_gender = dataset["Gender"].mode()[0]
```

```python
dataset["Gender"].value_counts()
```

```python
dataset["Gender"].fillna(mode_gender, inplace=True)
dataset.isnull().sum()
```

```python
#same as above, just changed syntax
dataset.fillna({"Gender":mode_gender}, inplace=True)
```

```python
dataset.isnull().sum()
```

```python
# to fill all my object dtypes with mode
dataset.select_dtypes(include="object")
```

```python
# we can see all the object datatype that has null val and now we will fill them with mode val
dataset.select_dtypes(include="object").isnull().sum()
```

```python
#this gives me the column names for those object dtypes 
dataset.select_dtypes(include="object").columns
```

```python
for i in dataset.select_dtypes(include="object").columns:
    # dataset[i].fillna(dataset[i].mode()[0],inplace= True)
    dataset.fillna({i:dataset[i].mode()[0]}, inplace =True)
```
