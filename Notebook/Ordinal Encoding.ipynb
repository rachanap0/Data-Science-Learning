### Ordinal Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
df.head()
```

```python
ord_data = [["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = ord_data) #by default, the category is auto which means the number will be assigned based on the alphabets.
#here we want the order to be based on our size i.e. s,m,l,xl so we passed ord_data to categories.
```

```python
df["Size_en"] = oe.fit_transform(df[["Size"]])
df
```

```python
#we can assign any random number, doesn't have to start with 0
ord_data_map = {"s":0,"m":1,"l":2,"xl":3}
```

```python
df["Size_mapped_en"] = df["Size"].map(ord_data_map)
```

```python
dataset = pd.read_csv("Loan.csv")
dataset.head()
```

```python
dataset.fillna({"Property_Area":dataset["Property_Area"].mode()[0]})
```

```python
dataset["Property_Area"].unique()
```

```python
#Encoding the Property_Area column using OrdinalEncoder
en_data_ord = [['Rural', 'Semiurban','Urban']]
```

```python
oen = OrdinalEncoder(categories=en_data_ord)
```

```python
dataset["Property_Area"] = oen.fit_transform(dataset[["Property_Area"]]).flatten()
```
